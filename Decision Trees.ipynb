## Training and Visualizing a Decision Tree

### Classification

Training of Decision Tree

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame=True)
X_iris = iris.data[["petal length (cm)", "petal width (cm)"]].values
y_iris = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X_iris, y_iris)
```

Visualization of Decision Tree

```python
from sklearn.tree import export_graphviz
from graphviz import Source

export_graphviz(
    tree_clf,
    out_file="iris_tree.dot",
    feature_names=["petal length (cm)", "petal width (cm)"],
    class_names = iris.target_names,
    rounded=True,
    filled=True
)

Source.from_file("iris_tree.dot")
```

### Estimating Class Probabilities

Suppose a flower whose petals are 5cm long and 1.5cm wide is found, the probability of the type of flower can be determined using Decision Tree.

```python
tree_clf.predict_proba([[5, 1.5]]).round(3)
```

The output tells us about the following probabilities:
* 0% for *Iris Setosa*
* 90.7% for *Iris Versicolor*
* 9.3% for *Iris Virginica*

### Regression

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

np.random.seed(42)
X_quad = np.random.rand(200, 1) - 0.5
y_quad = X_quad ** 2 + 0.025 * np.random.randn(200, 1)

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X_quad, y_quad)
```

```python
export_graphviz(
    tree_reg,
    out_file="reg_tree.dot",
    feature_names=["x1"],
    rounded=True,
    filled=True
)
Source.from_file("reg_tree.dot")
```
